```python
import os
import pandas as pd

# ê²½ë¡œ ì„¤ì •
base_path = "/Users/yiji/Desktop/ybigta/project/êµìœ¡ ê³µëª¨ì „/Contest_edudata/Database/schoolinfo/summary/combined_summary"

# ê²°ê³¼ë¥¼ ëˆ„ì í•  ë¦¬ìŠ¤íŠ¸
results = []

# ì§€ì—­ë³„ ë””ë ‰í† ë¦¬ ìˆœíšŒ
for region in os.listdir(base_path):
    region_path = os.path.join(base_path, region)
    
    # í´ë”ì¸ì§€ í™•ì¸ (íŒŒì¼ëª…ë“¤ê³¼ êµ¬ë¶„)
    if os.path.isdir(region_path):
        target_file = os.path.join(region_path, "combined_ì˜ˆì‚°_ì„¸ì¶œ_í•©ê³„ìš”ì•½.csv")
        
        if os.path.exists(target_file):
            df = pd.read_csv(target_file)
            # ì›í•˜ëŠ” ì»¬ëŸ¼ë§Œ ì¶”ì¶œ + í•„í„°ë§
            df_filtered = df[df["íŒŒì¼ëª…"] == "2023_í•©ê³„"][["íŒŒì¼ëª…", "í•™êµ_ì¼ë°˜ìš´ì˜", "í•™êµ_ì‹œì„¤í™•ì¶©", "í•™êµ_ì¬ë¬´í™œë™"]]
            # ì§€ì—­ëª… ì¶”ê°€
            df_filtered["ì§€ì—­"] = region
            results.append(df_filtered)

# ê²°ê³¼ í•©ì¹˜ê¸°
final_df = pd.concat(results, ignore_index=True)

# ê²°ê³¼ ì €ì¥ (ì„ íƒ)
final_df.to_csv("ì§€ì—­ë³„_ì˜ˆì‚°_ì„¸ì¶œ_í•©ê³„ìš”ì•½_2023.csv", index="íŒŒì¼ëª…")

# ê²°ê³¼ í™•ì¸
print(final_df)
```

```python
import pandas as pd

file_path = "news_report.csv"

# ğŸš¨ êµ¬ë¶„ì ìë™ ì„¤ì •: íƒ­ìœ¼ë¡œ êµ¬ë¶„
df = pd.read_csv(file_path, sep="\t")

# ì—´ ì´ë¦„ ê³µë°± ì œê±°
df.columns = df.columns.str.strip()

# ì—´ ì´ë¦„ í™•ì¸
print("ğŸ“Œ ì—´ ì´ë¦„ ëª©ë¡:", df.columns.tolist())

# ì—´ ì´ë¦„ ì§ì ‘ ì§€ì •
ì˜ˆì‚°ê¸ˆì•¡_ì—´ = "ì˜ˆì‚°ê¸ˆì•¡"
ë‹¤ìŒì˜ˆì‚°_ì—´ = "ë‹¤ìŒì—°ë„ì˜ˆì‚°"
ë¹„ìœ¨_ì—´ = "ë¹„ìœ¨"
ë‹¤ìŒë¹„ìœ¨_ì—´ = "ë‹¤ìŒì—°ë„ë¹„ìœ¨"

# ìˆ«ìí˜• ë³€í™˜
df[ì˜ˆì‚°ê¸ˆì•¡_ì—´] = pd.to_numeric(df[ì˜ˆì‚°ê¸ˆì•¡_ì—´], errors="coerce")
df[ë‹¤ìŒì˜ˆì‚°_ì—´] = pd.to_numeric(df[ë‹¤ìŒì˜ˆì‚°_ì—´], errors="coerce")

# ì „ì²´ í•©
total_current = df[ì˜ˆì‚°ê¸ˆì•¡_ì—´].sum()
total_next = df[ë‹¤ìŒì˜ˆì‚°_ì—´].sum()

# ë¹„ìœ¨ ê³„ì‚°
df[ë¹„ìœ¨_ì—´] = (df[ì˜ˆì‚°ê¸ˆì•¡_ì—´] / total_current * 100).round(2)
df[ë‹¤ìŒë¹„ìœ¨_ì—´] = (df[ë‹¤ìŒì˜ˆì‚°_ì—´] / total_next * 100).round(2)

# ì €ì¥
df.to_csv("news_report_with_ratio.csv", index=False)
print("âœ… ë¹„ìœ¨ ê³„ì‚° ë° ì €ì¥ ì™„ë£Œ!")
```
