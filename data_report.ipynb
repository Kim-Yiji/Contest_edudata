```python
import os
import pandas as pd

# 경로 설정
base_path = "/Users/yiji/Desktop/ybigta/project/교육 공모전/Contest_edudata/Database/schoolinfo/summary/combined_summary"

# 결과를 누적할 리스트
results = []

# 지역별 디렉토리 순회
for region in os.listdir(base_path):
    region_path = os.path.join(base_path, region)
    
    # 폴더인지 확인 (파일명들과 구분)
    if os.path.isdir(region_path):
        target_file = os.path.join(region_path, "combined_예산_세출_합계요약.csv")
        
        if os.path.exists(target_file):
            df = pd.read_csv(target_file)
            # 원하는 컬럼만 추출 + 필터링
            df_filtered = df[df["파일명"] == "2023_합계"][["파일명", "학교_일반운영", "학교_시설확충", "학교_재무활동"]]
            # 지역명 추가
            df_filtered["지역"] = region
            results.append(df_filtered)

# 결과 합치기
final_df = pd.concat(results, ignore_index=True)

# 결과 저장 (선택)
final_df.to_csv("지역별_예산_세출_합계요약_2023.csv", index="파일명")

# 결과 확인
print(final_df)
```

# 결과 보고서 뉴스 예산 계산

```python
import pandas as pd

file_path = "news_report.csv"

# 🚨 구분자 자동 설정: 탭으로 구분
df = pd.read_csv(file_path, sep="\t")

# 열 이름 공백 제거
df.columns = df.columns.str.strip()

# 열 이름 확인
print("📌 열 이름 목록:", df.columns.tolist())

# 열 이름 직접 지정
예산금액_열 = "예산금액"
다음예산_열 = "다음연도예산"
비율_열 = "비율"
다음비율_열 = "다음연도비율"

# 숫자형 변환
df[예산금액_열] = pd.to_numeric(df[예산금액_열], errors="coerce")
df[다음예산_열] = pd.to_numeric(df[다음예산_열], errors="coerce")

# 전체 합
total_current = df[예산금액_열].sum()
total_next = df[다음예산_열].sum()

# 비율 계산
df[비율_열] = (df[예산금액_열] / total_current * 100).round(2)
df[다음비율_열] = (df[다음예산_열] / total_next * 100).round(2)

# 저장
df.to_csv("news_report_with_ratio.csv", index=False)
print("✅ 비율 계산 및 저장 완료!")
```
